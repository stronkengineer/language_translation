```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, plot_confusion_matrix
from tensorflow.keras.models import load_model
import matplotlib.pyplot as plt
from sklearn.model_selection import learning_curve
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import ShuffleSplit

# Load the best model
best_model = load_model('checkpoint_sg.keras')
trained_embeddings = best_model.get_weights()[0]

num_classes = len(languages)
labels = np.random.randint(0, num_classes, size=len(trained_embeddings))

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(np.array(trained_embeddings), np.array(labels), test_size=0.2, random_state=42)

# Initialize and train the decision tree classifier with max_depth
max_depth = 10  # Set the desired maximum depth
decision_tree = DecisionTreeClassifier(max_depth=max_depth)
decision_tree.fit(X_train, y_train)

# Make predictions
y_pred = decision_tree.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Decision Tree Accuracy: {accuracy}")

# Classification Report
class_report = classification_report(y_test, y_pred)
print("Classification Report:")
print(class_report)

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
plt.imshow(conf_matrix, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.colorbar()

tick_marks = np.arange(num_classes)
plt.xticks(tick_marks, languages, rotation=45)
plt.yticks(tick_marks, languages)

for i in range(num_classes):
    for j in range(num_classes):
        plt.text(j, i, str(conf_matrix[i, j]),
                 horizontalalignment="center",
                 color="white" if conf_matrix[i, j] > conf_matrix.max() / 2 else "black")

plt.ylabel('True language')
plt.xlabel('Predicted language')
plt.tight_layout()
plt.show()



# Learning Curve
def plot_learning_curve(estimator, title, X, y, ylim=None, cv=None, n_jobs=None, train_sizes=np.linspace(.1, 1.0, 5)):
    plt.figure()
    plt.title(title)
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Training examples")
    plt.ylabel("Score")
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes)
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    plt.grid()

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
             label="Cross-validation score")

    plt.legend(loc="best")
    return plt


title = "Learning Curve (Decision Tree)"
cv = ShuffleSplit(n_splits=10, test_size=0.2, random_state=42)
plot_learning_curve(decision_tree, title, X_train, y_train, cv=cv, n_jobs=-1)

plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, plot_confusion_matrix
from sklearn.model_selection import train_test_split, learning_curve, ShuffleSplit
import matplotlib.pyplot as plt
import nltk

from gensim.scripts.glove2word2vec import glove2word2vec
from gensim.models import KeyedVectors

# Path to the GloVe file
glove_file = 'glove.6B.50d.txt'  # Update with the actual file path

# Convert GloVe format to word2vec format
word2vec_output_file = 'glove.6B.50d.word2vec.txt'
glove2word2vec(glove_file, word2vec_output_file)

# Load GloVe embeddings into memory using KeyedVectors
glove_model = KeyedVectors.load_word2vec_format(word2vec_output_file, binary=False)

# Assuming 'selected_rows' is your DataFrame containing 'Text' and 'language' columns
texts = selected_rows['Text'].tolist()
labels = selected_rows['language'].tolist()

# Function to calculate text embeddings using GloVe
def text_embedding(text, model):
    word_embeddings = []
    for word in text.split():  # Assuming text is a string of words
        if word in model:
            word_vector = model[word]
            word_embeddings.append(word_vector)
    
    if word_embeddings:
        word_embeddings = np.array(word_embeddings)  # Convert to NumPy array
        return np.mean(word_embeddings, axis=0)
    else:
        return np.zeros(model.vector_size)

# Generate text embeddings using GloVe
X = np.array([text_embedding(text, glove_model) for text in texts])

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.2, random_state=42)

# Initialize and train the Random Forest classifier
random_forest = RandomForestClassifier(n_estimators=120, max_depth=5)
random_forest.fit(X_train, y_train)

# Evaluate the model
y_pred = random_forest.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(10, 8))
plot_confusion_matrix(random_forest, X_test, y_test, cmap=plt.cm.Blues, display_labels=np.unique(labels))
plt.xticks(rotation=45)  # Rotate x-axis labels by 45 degrees
plt.title('Confusion Matrix')
plt.show()

# Learning Curve
def plot_learning_curve(estimator, title, X, y, ylim=None, cv=None, n_jobs=None, train_sizes=np.linspace(.1, 1.0, 5)):
    plt.figure()
    plt.title(title)
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Training examples")
    plt.ylabel("Score")
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes)
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    plt.grid()

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
             label="Cross-validation score")

    plt.legend(loc="best")
    return plt

title = "Learning Curve (Random Forest)"
cv = ShuffleSplit(n_splits=10, test_size=0.2, random_state=42)
plot_learning_curve(random_forest, title, X_train, y_train, cv=cv, n_jobs=-1)

plt.show()
```
